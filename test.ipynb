```python
import torch
from moco.loader import TwoCropsTransform
from moco.builder import MoCo
import torchvision.models as models
import torchvision.datasets as datasets
import os
import torchvision.transforms as transforms
```

```python
model = MoCo(
        models.__dict__["resnet50"],
        4096,
        128,
        0.07,
        False,
    )
```

```python
statedict = torch.load("/mnt/nvme0n1p2/Moco-MAC/SOCOLA/checkpoint_0026.pth.tar")
model.load_state_dict(statedict["state_dict"])
img_encoder = model.encoder_img
```

```python
datadir = "/mnt/nvme0n1p2/CVPR"
normalize = transforms.Normalize(
        mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]
    )
augmentation = [
            transforms.RandomResizedCrop(224, scale=(0.2, 1.0)),
            transforms.RandomGrayscale(p=0.2),
            transforms.ColorJitter(0.4, 0.4, 0.4, 0.4),
            transforms.RandomHorizontalFlip(),
            transforms.ToTensor(),
            normalize,
        ]
transform = transforms.Compose(augmentation)
```

```python
from PIL import Image

image = Image.open("/mnt/nvme0n1p2/CVPR/ILSVRC/imagenet/n01443537/n01443537_2.JPEG")
image2 = Image.open("/mnt/nvme0n1p2/CVPR/ILSVRC/imagenet/n01514668/n01514668_19.JPEG")
```

```python

input_tensor = transform(image).unsqueeze(0) 
input_tensor2 = transform(image2).unsqueeze(0)
```

```python
with torch.no_grad():
    output = img_encoder(input_tensor)
    output2 = img_encoder(input_tensor2)
output.shape
```
